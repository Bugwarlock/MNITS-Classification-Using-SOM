```python
import numpy as np
w11 = 0.2
w12 = 0.3
w21 = 0.4
w22 = 0.5
b0 = 0.1
b1 = 0.95
inp = np.zeros((2,2))
inp[0,0] = 1
inp[0,1] = -1
inp[1,0] = -1
inp[1,1] = 1
x = np.zeros(2)
x[0] = -1
x[1] = 1
z1 = 0
z2 = 0
alpha = 0.5
# error = np.full(2,1)
error=1
counter = 0
# while(np.count_nonzero(error) != 0):
while(error != 0):
    error=0
    for i in range(2):
        for j in range(2):
            z1 = ((w11*x[i])+(w21*x[j])+b0)
            z2 = ((w12*x[i])+(w22*x[j])+b1)
            if (z1 >0):
                h1 = 1
            else:
                h1 = -1
            if (z2 >0):
                h2 = 1
            else:
                h2 = -1
            ht = 0.5+ (0.5*h1)+(0.5*h2)
            if (ht >0):
                ht = 1
            else:
                ht = -1
            if(inp[i,j] != ht):
                error +=  1
            if((inp[i,j] != ht) and (inp[i,j]==-1)):
                if(z1>0):
                    w11 = w11 + (alpha*(inp[i,j]-z1)*x[i])
                    w21 = w21 + (alpha*(inp[i,j]-z1)*x[j])
                    b0 = b0 + (alpha*(inp[i,j]-z1))
                if(z2>0):
                    w12 = w12 + (alpha*(inp[i,j]-z2)*x[i])
                    w22 = w22 + (alpha*(inp[i,j]-z2)*x[j])
                    b1 = b1 + (alpha*(inp[i,j]-z2))
            if((inp[i,j] != ht) and (inp[i,j]==1)):
                if(abs(z1)<abs(z2)):
                    w11 = w11 + (alpha*(inp[i,j]-z1)*x[i])
                    w21 = w21 + (alpha*(inp[i,j]-z1)*x[j])
                    b0 = b0 + (alpha*(inp[i,j]-z1))
                elif(abs(z1)>abs(z2)):
                    w12 = w12 + (alpha*(inp[i,j]-z2)*x[i])
                    w22 = w22 + (alpha*(inp[i,j]-z2)*x[j])
                    b1 = b1 + (alpha*(inp[i,j]-z2))
                else:
                    w11 = w11 + (alpha*(inp[i,j]-z1)*x[i])
                    w21 = w21 + (alpha*(inp[i,j]-z1)*x[j])
                    b0 = b0 + (alpha*(inp[i,j]-z1))
                    w12 = w12 + (alpha*(inp[i,j]-z2)*x[i])
                    w22 = w22 + (alpha*(inp[i,j]-z2)*x[j])
                    b1 = b1 + (alpha*(inp[i,j]-z2))
        counter +=1
print('error',error)
print(counter)
print(w11)
print(w21)
print(w12)
print(w22)
print(b0)
print(b1)
```
