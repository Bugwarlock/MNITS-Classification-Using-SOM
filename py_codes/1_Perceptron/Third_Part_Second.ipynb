```python
import numpy as np
inp1 = np.full(81,-1)
inp1[4]=1
inp1[10]=1
inp1[36]=1
inp1[44]=1
inp1[64]=1
inp1[76]=1
inp1[79]=1
inp2 = np.full(81,-1)
inp2[23]=1
inp2[40]=1
inp2[48]=1
w1 = np.random.rand(81)
w2 = np.random.rand(81)
b1 = np.random.rand()
b2 = np.random.rand()
alpha = 0.5
error1 = np.full(2,1)
error2 = np.full(2,1)
counter = 0
# for x in range(11):
while((np.count_nonzero(error1) != 0) or (np.count_nonzero(error2) !=0)):
    #for first input set
    summ1 = 0
    summ2 = 0
    for i in range(81):
        summ1 += (w1[i]*inp1[i])+b1
        summ2 += (w2[i]*inp1[i])+b2
    if (summ1 > 0):
        h1 = 1
    else:
        h1 = -1
    if (summ2 > 0):
        h2 = 1
    else:
        h2 = -1
    error1[0] = (1-h1)
    error1[1] = (-1-h2)
    if (h1 != 1):
        for i in range(81):
            w1[i] = w1[i] + (1*inp1[i]*alpha)
        b1 = b1 + 1*alpha
    if (h2 != -1):
        for i in range(81):
            w2[i] = w2[i] - (1*inp1[i]*alpha)
        b2 = b2 - 1*alpha
    #for second input set
    summ1 = 0
    summ2 = 0
    for i in range(81):
        summ1 += (w1[i]*inp2[i])+b1
        summ2 += (w2[i]*inp2[i])+b2
    if (summ1 > 0):
        h1 = 1
    else:
        h1 = -1
    if (summ2 > 0):
        h2 = 1
    else:
        h2 = -1
    error2[0] = (-1-h1)
    error2[1] = (1-h2)
    if (h1 != -1):
        for i in range(81):
            w1[i] = w1[i] - (inp2[i]*1*alpha)
        b1 = b1 - 1*alpha
    if (h2 != 1):
        for i in range(81):
            w2[i] = w2[i] + (inp2[i]*1*alpha)
        b2 = b2 + 1*alpha
    counter +=1        
print(error1)
print(error2)
print('w1' ,w1)
print('w2' ,w2)
print('b1' ,b1)
print('b2' ,b2)
print('epoch' ,counter)
```
